```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
base_path = ''
```

```python
# a 마트 gds+list+dtl
amartsum = pd.read_csv(base_path + 'amartsum.csv')
amartsum
```

```python
amartsum.SAL_DT = amartsum.SAL_DT.astype('str')
```

```python
amartsum.REAL_SAL_DT = amartsum.REAL_SAL_DT.astype('str')
```

```python
amartsum.REAL_SAL_TTM = amartsum.REAL_SAL_TTM.astype('str')
```

```python
amartsum.SAL_DT = pd.to_datetime(amartsum.SAL_DT)
```

```python
t1 = amartsum['REAL_SAL_TTM'].astype(str).str.zfill(6)
amartsum['REAL_SAL_TTM'] = pd.to_datetime(t1, format='%H%M%S').dt.time

print (amartsum)
```

```python
amartsum['hour'] = pd.to_datetime(amartsum['REAL_SAL_TTM'], format='%H:%M:%S').dt.hour
amartsum
```

```python
amartsum['month'] = pd.to_datetime(amartsum['SAL_DT'], format='%Y-%m-%d').dt.month
amartsum
```

```python
amartsum['day'] = pd.to_datetime(amartsum['SAL_DT'], format='%Y-%m-%d').dt.weekday
amartsum
```
