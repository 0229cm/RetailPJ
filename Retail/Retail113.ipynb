```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

base_path = ''
```

```python
df = pd.read_csv(base_path + 'cat1014.csv')
df
```

```python
df1 = df[df['GDS_NM'] == '대파1단']
df1
```

```python
df1['SALE_QTY'].sum()
```

```python
df.SAL_DT = df.SAL_DT.astype('str')
```

```python
df.SAL_DT = pd.to_datetime(df.SAL_DT)
```

```python
df = df.astype({'SALE_QTY' : 'int'})
```

```python
df = df.astype({'SALE_PRC' : 'int'})
```

```python
df = df.astype({'SALE_AMT' : 'int'})
```

```python
df['month'] = pd.to_datetime(df['SAL_DT'], format='%Y-%m-%d').dt.month
```

```python
df['weekday'] = pd.to_datetime(df['SAL_DT'], format='%Y-%m-%d').dt.weekday
```

```python
df1 = df[df['SALE_QTY'] > 0]
```

```python
df2 = df1[df1['SALE_PRC'] > 0]
```

```python
df3 = df2.groupby('GDS_NM').sum()
df3
```

```python
df4 = df3.reset_index()
df4
```

```python
df4.sort_values('SALE_QTY', ascending=False)
```

```python
df5 = df4[df4['SALE_QTY'] > 365]
df5
```

```python
df6 = df5[['GDS_NM', 'SALE_QTY']]
df6
```

```python
df7 = df6.sort_values('SALE_QTY', ascending=False)
df7
```
