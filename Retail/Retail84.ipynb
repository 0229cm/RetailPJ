```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
base_path = ''
```

```python
df = pd.read_csv(base_path + 'cat1014.csv')
df
```

```python
df.info()
```

```python
df = df.astype({'SALE_QTY' : 'int'})
```

```python
df = df.astype({'SALE_AMT' : 'int'})
```

```python
df = df.astype({'SAL_DT' : 'str'})
```

```python
df['SAL_DT'] = pd.to_datetime(df['SAL_DT'])
```

```python
df.info()
```

```python
df
```

```python
df['GDS_NM'].value_counts()
```

```python
daepa = df[df['GDS_NM'] == '대파1단']
daepa
```

```python
data = daepa[['SAL_DT', 'SALE_QTY', 'SALE_PRC', 'SALE_AMT']]
data
```

```python
data = data[['SAL_DT', 'SALE_QTY']]
data
```

```python
data.sort_values('SAL_DT')
```

```python
data = data.groupby('SAL_DT').sum()
data
```

```python
data.index
```

```python
plt.figure(figsize = (20,10))
sns.lineplot(x = data.index, y = data['SALE_QTY'])
```

```python
data.plot(figsize = (20,10))
```

```python
data.resample('MS').mean() # 월별
```

```python
data = data.resample('MS').mean()
```

```python
data.plot(figsize = (20,10))
```

```python
from fbprophet import Prophet
```

```python
data
```

```python
df = data.reset_index()
```

```python
df
```

```python
df.columns = ['ds','y']
```

```python
df
```

```python
model = Prophet()
```

```python
model.fit(df)
```

```python
future = model.make_future_dataframe(periods = 12, freq = 'MS')
```

```python
future
```

```python
model.predict(future)
```

```python
pred = model.predict(future)
```

```python
pred = pred[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
pred
```

```python
model.plot(pred)
plt.show()
```
