```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
base_path = ''
```

```python
# 제품목록 gds_list_0, gds_list_1
gds_2 = pd.read_csv(base_path + 'gds2.csv')

# 제품 카테고리 목록: gdsclss_list_0, gdsclss_list_1
gdsclss_2 = pd.read_csv(base_path + 'gdsclss_2.csv')

# 매출 데이터 마스터(장바구니 단위): tr_list_0, tr_list_1
tr_list_2 = pd.read_csv(base_path + 'tr_list_2.csv')

# 매출 제품 데이터 상세: tr_dtl_list_0, tr_dtl_list_1
tr_dtl_list_2 = pd.read_csv(base_path + 'tr_dtl_list_2.csv')
```

```python
c_mer = pd.merge(tr_dtl_list_2[['SAL_DT','SEQ_NO', 'PLU_CD']],gds_2[['PLU_CD', 'GDS_NM']])
c_mer
```

```python
dtl_copy = tr_dtl_list_2.copy()
```

```python
dtl_mer = pd.merge(dtl_copy, gds_2, how='left', on='PLU_CD')
dtl_mer
```

```python
dtl_mer.drop(['CO_CD', 'STR_CD_y', 'GDS_CLSS_CD_x','SGDS_DC_PRC', 'SUBTOT_DC_AMT', 'EVT_DC_AMT', 'CPON_DC_AMT','GDS_CPON_DC_AMT','BASKET_CPON_DC_AMT'], axis=1, inplace=True)
dtl_mer
```

```python
dtl_mer.reindex(columns=['SAL_DT', 'STR_CD_x', 'POS_NO', 'TRAN_NO', 'SEQ_NO', 'DESI_CNCL_TP_NM',
       'SALE_RTN_SGN', 'PLU_CD', 'GDS_NM', 'GDS_CLSS_CD_y', 'SALE_QTY', 'SALE_PRC', 'SALE_AMT',
       'SGDS_DC_TP_NM', 'SCALES_GDS_TP_NM', 
       'GDS_TP_NM', 'VEN_CD'])
```

```python
dtl_mer_copy = dtl_mer.copy()
```

```python
dtl_mer_copy.SAL_DT = dtl_mer_copy.SAL_DT.astype('str')
dtl_mer_copy.SAL_DT
```

```python
dtl_mer_copy.SAL_DT = pd.to_datetime(dtl_mer_copy.SAL_DT)
# dtl_mer_copy.SAL_DT.dt.strftime('%Y/%m/%d')
dtl_mer_copy
```

```python
dtl_gr = dtl_mer_copy.groupby(['SAL_DT', 'TRAN_NO'])['GDS_NM'].apply(list)
dtl_gr
```

```python
dtl_gr = dtl_gr.to_frame()
dtl_gr
```

```python
dtl_re = dtl_gr.reset_index().rename(columns={"index":"TRAN_NO"})
dtl_re
```

```python
dtl_NM_3 = dtl_re['GDS_NM'].to_frame()
dtl_NM_3
```

```python
dtl_NM_3.to_csv('dtl_NM_3.csv', index=False)
```
