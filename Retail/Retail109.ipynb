```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
base_path = ''
```

```python
df = pd.read_csv(base_path + 'df1018.csv') # 1015 인데 1005로 저장 잘못했네..
df
```

```python
df.info()
```

```python
data = df[['SAL_DT', 'SALE_QTY', 'SALE_PRC', 'SALE_AMT']]
data
```

```python
data['SAL_DT'] = pd.to_datetime(data['SAL_DT'])
data
```

```python
data = data[['SAL_DT', 'SALE_AMT']]
data
```

```python
data.sort_values('SAL_DT')
```

```python
data = data.groupby('SAL_DT').sum()
data
```

```python
data.index
```

```python
plt.figure(figsize = (20,10))
sns.lineplot(x = data.index, y = data['SALE_AMT'])
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])
```

```python
data.resample('MS').mean() # 월별
```

```python
from fbprophet import Prophet
```

```python
data
```

```python
df = data.reset_index()
df
```

```python
df.columns = ['ds','y']
df
```

```python
model = Prophet()
```

```python
model.fit(df)
```

```python
future = model.make_future_dataframe(periods = 12, freq = 'MS')
# 12개월 / 월 단위 리셋을 나타내는 파라미터 (periods = 12, freq = 'MS')
```

```python
model.predict(future)
```

```python
pred = model.predict(future)
```

```python
pred = pred[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
pred
```

```python
model.plot(pred)
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])
plt.show()
```
