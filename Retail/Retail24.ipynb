```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
base_path = ''
pd.set_option('display.max_row', 500)
```

```python
test = pd.read_csv(base_path + 'test0928.csv')
test
```

```python
test1 = test[['GDS_NM', 'hour', 'month', 'day']]
test1
```

```python
test1['GDS_NM']= test1['GDS_NM'].str.replace(pat=r'[^가-힣A-Z0-9]',repl=r' ',regex=True)
test1['GDS_NM']
```

```python
test1
```

```python
texts = test1['GDS_NM']
texts
```

```python
from hanspell import spell_checker
```

```python
from tqdm import tqdm
```

```python
spell_check = []
for i in tqdm(range(len(texts))):
    try:
        result_train = spell_checker.check(texts[i])
        texts[i] = result_train.checked
        spell_check.append(texts[i])
    except:
        spell_check.append("error")
        
spell_check[:10]
```
